```
import altair as alt
import pandas as pd
datasetURL="https://raw.githubusercontent.com/theoliao1998/Infomation-Visualization/master/Individual%20Assignment/a1-burtin-long.csv?token=AM26APNYRBHFGNFCY7QEUQK6ITC34"
data=pd.read_csv(datasetURL, encoding="latin-1")
data.columns = ["_".join(c.split()) for c in data.columns]
```

```
data
```

```
main = alt.Chart(data).mark_circle(size=800,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).encode(
    x="Antibiotic:N",
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000])),
    color="Gram_Staining_",
).properties(width=600)

main
```

```
box1 = alt.Chart(data).transform_filter(
    alt.datum.Gram_Staining == 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="orange",size=50,opacity=0.3).encode(
    x="Antibiotic:N",
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000])),
).properties(width=600)

box2 = alt.Chart(data).transform_filter(
    alt.datum.Gram_Staining != 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="blue",size=50,opacity=0.3).encode(
    x="Antibiotic:N",
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000])),
).properties(width=600)


box1
```

```
main_S = alt.Chart(data).mark_circle(size=1400,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic != "Neomycin",
).transform_filter(
    alt.datum.Antibiotic != "Penicilin",
).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labelAngle=0)),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
    color=alt.Color("Gram_Staining", scale=alt.Scale(domain=['positive', 'negative'],scheme='set1')),
).properties(width=70)

pos_S = alt.Chart(data).mark_circle(size=800,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic != "Neomycin",
).transform_filter(
    alt.datum.Antibiotic != "Penicilin",
).transform_filter(
    alt.datum.Gram_Staining == 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="red",size=40).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False),title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None),
).properties(width=60)

neg_S = alt.Chart(data).mark_circle(size=800,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic != "Neomycin",
).transform_filter(
    alt.datum.Antibiotic != "Penicilin",
).transform_filter(
    alt.datum.Gram_Staining != 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="#377eb8",size=40).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False), title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
).properties(width=60)

S = alt.hconcat(alt.hconcat(pos_S, main_S,spacing=0),neg_S,spacing=0)
```

```
main_P = alt.Chart(data).mark_circle(size=1400,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "Penicilin",
).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labelAngle=0),title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
    color=alt.Color("Gram_Staining", scale=alt.Scale(domain=['positive', 'negative'],scheme='set1')),
).properties(width=70)

pos_P = alt.Chart(data).mark_circle(size=600,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "Penicilin",
).transform_filter(
    alt.datum.Gram_Staining == 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="red",size=40).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False), title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=alt.Axis(grid=False,),title="1/MIC"),
).properties(width=60)

neg_P = alt.Chart(data).mark_circle(size=800,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "Penicilin",
).transform_filter(
    alt.datum.Gram_Staining != 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="#377eb8",size=40).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False), title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
).properties(width=60)

P = alt.hconcat(alt.hconcat(pos_P, main_P,spacing=0),neg_P,spacing=0)
```

```
main_N = alt.Chart(data).mark_circle(size=1400,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "Neomycin",
).encode(
    x=alt.X("Antibiotic:N",title="",axis=alt.Axis(labelAngle=0)),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
    color=alt.Color("Gram_Staining", scale=alt.Scale(domain=['positive', 'negative'],scheme='set1')),
).properties(width=70)

pos_N = alt.Chart(data).mark_circle(size=600,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "Neomycin",
).transform_filter(
    alt.datum.Gram_Staining == 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="red",size=40).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False), title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None),
).properties(width=60)

neg_N = alt.Chart(data).mark_circle(size=600,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "Neomycin",
).transform_filter(
    alt.datum.Gram_Staining != 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="#377eb8",size=40).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False), title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
).properties(width=60)

N = alt.hconcat(alt.hconcat(pos_N, main_N,spacing=0),neg_N,spacing=0).resolve_scale(y="shared")
N
```

```
alt.hconcat(alt.hconcat(P, S,spacing=0),N,spacing=0).resolve_scale(y="shared").properties(
    title='Effectiveness of Antibiotics on Gram-positive and Gram-negative Bacteria'
).configure_view(strokeOpacity=0)
```

```
space = alt.Chart(data).mark_circle(size=600,opacity=0.7).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).transform_filter(
    alt.datum.Antibiotic == "S",
).transform_filter(
    alt.datum.Gram_Staining != 'positive',
).transform_calculate(
    invert_MIC = "1.0/datum.MIC",
).mark_boxplot(color="#377eb8",size=30).encode(
    x=alt.X("Antibiotic:N",axis=alt.Axis(labels=False), title=""),
    y=alt.Y("invert_MIC:Q",scale=alt.Scale(type="log",domain=[0.0001, 10000]),axis=None, title=""),
).properties(width=50)
```

```
p = alt.hconcat(P,space,spacing=0)
n = alt.hconcat(space,N,spacing=0)

alt.hconcat(alt.hconcat(p, S,spacing=0),n,spacing=0).resolve_scale(y="shared").properties(
    title='Effectiveness of Antibiotics on Gram-positive and Gram-negative Bacteria'
).configure_view(strokeOpacity=0)
```
